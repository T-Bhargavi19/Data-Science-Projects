```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
car_df=pd.read_csv("Car_Purchasing_Data.csv",encoding='ISO-8859-1')
```

```python
car_df.head()
```

```python
car_df= car_df.drop(['Customer Name','Customer e-mail','Country'], axis=1)
```

```python
car_df.head()
```

```python
car_df.corr()
```

```python
x=car_df.drop(['Car Purchase Amount'], axis=1)
```

```python
x.head()
```

```python
car_x=sm.add_constant(x)
```

```python
car_x.head()
```

```python
car_y= car_df['Car Purchase Amount']
```

```python
car_y.head()
```

```python
model=sm.OLS(car_y,car_x).fit()
model.summary()
```

```python
x.head()
```

```python
car_x2=x.drop(['Gender','Credit Card Debt'],axis=1)
```

```python
car_x2.head()
```

```python
car_x3=sm.add_constant(car_x2)
```

```python
model=sm.OLS(car_y,car_x3).fit()
model.summary()
```

```python
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as rmse
from sklearn.metrics import r2_score as r2
```

```python
print(f"MAE: {mae(car_y, model.predict())}")
```

```python
print(f"RMSE: {rmse(car_y, model.predict(), squared=False)}")
```
